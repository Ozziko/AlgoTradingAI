# Framework

## Definitions, assumptions
* There is a tradable **target** time series $y(t)$ that we wish to learn and predict in order to trade and make profits. For example, in this project the target is Google stock.
* There are **features** saved in the design matrix $X(t)$, that we will use to predict the target. Feature building is discussed under  *Pre-processing* section.
* The prediction must be **causal**: the target on a time $t$, $y(t=t')$, can only be predicted by past information, $X(t<t')$! Notice the difference from non-temporal ML, where $X^{(i)}$ is used to learn/predict  $y^{(i)}$.
* The trading strategy must be **feasable in practice, in real time**. 

## Algo-trading challenges
**Challanges originating in time-series learning:**
* **The main challenge in algo-trading: time dependence**.
  * **Learning long term dependencies:** unlike modeling statioanary, mean or simply frequent enough target dependancies on the inputs, it is hard to learn a hypothesis that "remembers" patterns for a long time since they occured. Goodfellow et al., *Deep Learning*, 2016: "the problem of learning long-term dependencies remains one of the main challenges in deep learning".
  * **Time dependent hypothesis**: even if the hypothesis captured well enough the target dependencies, unlike typical learning tasks such as image recognition - in algo-trading the hypothesis is  generally time dependent:
    * Different periods of the same length, even if long enough, can yield learnt hypotheses that substantially differ from one another. 
    * A learnt hypothesis has an **"expiry date"**, its validity can be short lived.
    * **Conclusions on the training/validation/testing period lengths are difficult to reach** - changing or shifting periods might change the target dependencies found in them, the learning "curves" are 2D surfaces - the train and test MSE as a function of the lengths of the train AND test periods.
* **Time has direction and continuity that should be conserved in the learning**, for example splitting to training and validation cannot be done by random sampling of the data.
  
**Challanges characterizing algo-trading in particular:**
* **The bottom line is more important than the learning itself**. *The goal in algo-trading is not necessarily to have the smallest error in predicting the target*. For example, if a prediction that the target price will increase by 10% in the next time step led to opening a long, and the price went up "only" by 5%, then the prediction was "bad" (relarive error of 100%) but the trading was profitable. An algorithm that misses the target, but in the right direction as in this example, can have consistently high error, and consisten profits.
* The goal in algo-trading is to be **profitable** and **consistent**, meaning that the trading strategy is profitable on the test period in a statistically significant way - the final profit is sufficiently higher than the std of the profit of trades along the way.

## Roadmap for this project
The learning algorithms that seem to best answer the challanges of time-series prediction found in algo-trading are **recurrent neural networks** (RNNs), and especially gated RNNs like LSTM that are designed to handle long time dependencies.

## A heuristic for my basic (non-AI) trading strategy 
My basic non-AI strategy - repeat each time a new prediction is made for the next time step:
* **go long (+1)** - if the target price is predicted to go up in a high enough confidence.
* **return to neutral (0)** - if the confidence for the prediction is not high enough.
* **go short (-1)** (if the target is shortable) - if the target price is predicted to go down in a high enough confidence.

Notes:
* **Confidence** - can be extracted from the learning algorithm, discussed below.
* Each position (long, neutral, short) is applied to all the existing portfolio, to have **exponential profits** (or losses).

---
Long & short in a nutshell:
* being long - having a positive amount of equity units, (for example, if the target is a stock - having a positive number of shares), profitable when the price increases relative to the price in which the long opened.
* being short - "the opposite of long", having a negative amount of equity units (=debt), profitable when the price decreases relative to the price in which the long opened. If needed, see an explanation about shorting in [Wikipedia](http://en.wikipedia.org/wiki/Short_(finance).


## Schemes for translating learning into trading predictions & decisions 

Even though the target in algo-trading is continuous, the trading decisions in my basic trading strategy described above are discrete: go long, return to neutral, or go short.

My basic trading strategy don't care to know on each time step the next absolute target price, but the relative price difference from current price, and the prediction confidence level.

**I present here two schemes for using learning to predict and make trading decisions:**


### My simple regression trading scheme
Predicting the target price by regression (shallow or deep), translating to relative price movements in %, then **translating to discrete trading decisions** -

If the next predicted relative price movement (%) is:
  * higher than confidence level - **go long**,
  * else if lower than the negative confidence level - **go short**,
  * else - **return to neutral**.
  
**Confidence level** can be: 
  * simply set to zero, meaning going long if next relative price movement (%) is positive, otherwise go short.
  * found as a hyper-parameter.
  * **inferred statistically**, as ~std of the training error, already known during test time, where $error(t)=predicted(t)-true(t)$. 
  
    For example, if the training error std is $\sigma=2\%$ and the confidence level is set to $\sigma=2\%$, then going long can be chosen only if predicted price movement is higher than confidence level, $predicted(t)>\sigma=2\%$. Loss will occure if the true price movement is negative, $true(t)<0\%$, which will cause $TestError(t)=predicted(t)_{>\sigma}-true(t)_{<0\%}>\sigma$. Assuming the test error is similar to the train error which is approximately Gaussian, the probability of loss is therefore $P(TestError>\sigma)\simeq{P(TrainError>\sigma)}\approx{1/2(1-0.683)}\approx{15.9\%}$

### My simple classification trading scheme
Calculating relative price movements (%) from target prices, **discretizing the ground truth target** into +1,0,-1 classes representing the ground truth trading decisions, then predicting the classes which directly represent the trading decisions to be taken!
  
Discretizing the target - if the target relative price movement (%) is:
  * higher than confidence level: **up (go long) class (+1)**,
  * else if lower than the negative confidence level - **down (go short) class (-1)**,
  * else - **neutral class (0)**.
  
**Confidence level** can be: 
  * simply set to zero, meaning positive relative price movement (%) is +1 class, otherwise -1 class.
  * found as a hyper-parameter.
  * **inferred statistically**, is a similar spirit as inferring the confidence level in regression: as ~std of the training % price movement distribution, already known during test time. Notice difference from the regression confidence (there it was ~std of the training *error*). 

### My classes structure
I constructed ***regression* and *classification* classes** to perform all the calculations required to go from data to predictions and predicted trading decisions.

Both classes lack training methods, which are defined in:
* ***regression_shallow* and *regression_deep* classes**, that inherit *OzML.regression*, and add methods relevant for shallow or deep learning.
* ***classification_shallow* and *classification_deep* classes**, that inherit *classification*, and add methods relevant for shallow or deep learning.

The final output of all classes - trading decisions, can be checked agains the test data to calculate the profits (with ***trading* class**).

### Building features while maintaining causaility
**All features must maintain causality:** the target on a time $t'$, $y(t=t')$, can only be learnt/predicted by past information, $X(t<t')$! 

However, the convention in ML is that each row of the design matrix $X$ contains the features to learn/predict the corresponding row of $y$. Therefore, this code will split the raw data in *data_df* into the the design matrix $X(t)$ and the target $y(t)$, shift the features one time step back relative to the target while keeping the index fixed, such that $X(t)=(x_0(t-\Delta t)),...,x_k(t-\Delta t)))$ can be used to learn/predict $y(t)$.

````
target_name='GOOGL Close'
include_target_history_as_features=True
# ------------------ end of inputs ------------------
(X_df,y_series)=OzML.basic_feature_building(data_df,target_name,
  attribute_names_list,include_target_history_as_features)

print('The target y_series first 5 rows:\n',y_series.iloc[:5])
print('\nThe design matrix X_df contains %d features (columns). Top 5 rows:\n'
     %len(X_df.columns))
X_df.head(5)
````

### Causality verification 
To verify that feature calculations don't violate causality, **I invented a simple verification: white noise regression**. There is a [sub-section](#scrollTo=IKJ4AvqLsoui) presenting it. In a nutshell: regression on completely random inputs taken from a distribution $N(0,\sigma^2)$ must yield a completely random prediction, and therefore error distributed as $N(0,\sigma^2)$ - or else cheating with information from the future (=breaking causality)!

### Linear regression
Linear regression is the most basic regression, but is it still advantageous:
* Forces a simple (maybe the simplest) hypothesis, that can be taken as a baseline model.
* Easy (maybe the easiest) to interpret, intuitive.
* Has no free hyper-parameters.
* Easy to regularize - L2 (Ridge), L1 (Lasso) or both (ElasticNet, in the next section).

It's a good place to start.

````
regression_obj=OzML.regerssion_shallow(train_test_split_obj)
regression_obj.train_skLinearRegression()
````

The first thing we can check - the learnt coefficients, mainly for comaprison with the next algorithms we will discuss:

````
regression_obj.visualize_coeff(max_coeff_to_show=25,attribute_names_list=attribute_names_list)
````

Having coefficients with a simple interpretation is an advantage of shallow learning, and coefficients of a linear hypothesis may have the simplest interpretation.

Interesting things to notice:
* Not suprisingly, the biggest coefficient belongs to the target 1 time-step history (the feature is $y(t-\Delta t)$), while the target, GOOGL Close, is $y(t)$).
* At first sight it might seem strange that the target depends inversely (negative coefficients) on APPL High, SPY Close, etc. - but this can be understood when noticing that APPL Close, SPY Close etc. are taken with positive and similar coefficients, meaning that **the target depends on "middle prices'**: APPL High-Close~Middle, SPY High-Close~Middle...
* **Volumes do not play a significant role** (at least in linear regression), which is not surprising given their high volatiliy that we observed. 

Now let's examine the prediction:

````
regression_obj.predict()
````

Notes on prediction statistics:
* Even though $MSE=mean(error^2)$ is the loss in linear regerssion, its units are $[error]^2=[target]^2$ (which is $\$^2$ in our case), so in my opinion it more intuitive to inspect $\sqrt{MSE}$.
* My trading schemes presented in this project translate the predictions into relative price movements, to make trading decisions. Therefore, the mean of the error is irrelevant, and $std(error)=\sqrt{mean((error-mean(error))^2)}$ should be used instead of $\sqrt{MSE}$ to asses errros.

### ElasticNet
ElasticNet is "nothing more" than a linear regression with L1-norm and L2-norm regularizations, however:
* I found the geometric interpretation of norm regularizations interesting.
* It can be used for **features post-selection** - by forcing a sparse hypothesis with minimal number of non-zero coefficients, it reveals what are the crucial features.

This comes with the expense of **adding two more free hyper-parameters to the learning** - the strengths of the two regularizatoins.

---
The optimization problem that ElasticNet solves: 

$min(1/2n_{samples}\times||y - Xw||^2_2+\alpha r||w||_1+1/2 \times\alpha(1-r||w||^2_2)$

* $0<\alpha$ ('alpha' parameter) is the regularization strength ($\alpha=0$ means no regularization - linear regression, but in practice it is not recommended).
* $0\leq r\leq 1$ ('l1_ratio') is the ratio bewteen L1 and L2 regularizations:
  * $r=1$ is a complete L1 regularization (Lasso),
  * $r=0$ is a complete L2 regularization (Ridge).

---
**Just to get a feeling about this algorithm we will use convenient regularization parameters I chose in advance. Of course that the proper way to find hyper-parameters is by cross validation** (see [Hyper-searching learning/trading parameters](#scrollTo=qe0a4pwoy-c9)).

````
regression_obj=OzML.regerssion_shallow(train_test_split_obj)
regression_obj.train_skElasticNet(l1_ratio=0.7,alpha=0.1)
regression_obj.visualize_coeff(max_coeff_to_show=25,normalization=True,bottom_margin=0.3)
````

Plotting the coefficients show how the ElasticNet regularization forced a sparse hypothesis with 4 significant coefficients, compared to ~18 in the (un-regularized) [linear regression](#scrollTo=VGfTenevwLtL).

Let's check the prediction:

````
regression_obj.predict()
regression_obj.plot_prediction(rolling_window=100);
````

The ElasticNet error std are a bit higher than those in linear regression.

The trading results:

````
pred_confidence_lebel=0
# ------------------ end of inputs ------------------
pred_test_position_series=regression_obj.decide_position(
        pred_confidence_lebel*regression_obj.train_err_std)
trading_obj=OzML.trading(pred_test_position_series,
                                train_test_split_obj.y_test_diff_percents_series)
trading_obj.plot_portfolio()
````

This is interesting. These results, with *pred_confidence_lebel=0*, are much better than those of the (un-regularized) linear regression with *pred_confidence_lebel=0*. However, persisting with the 2012-2015 period would have been too hard in a real situation. 

---
Satisfying results can be achieved with ElasticNet, when the hyper-parameters (regularization, train and test period lengths) are found by cross validation. This is treated in [Meta-learning](#scrollTo=I58MN7a6y3vS).

### Logistic classification
Logistic classification is the most basic classification. However, not only that it can be proved useful, **I find the derivation of multinomial logistic classification (softmax) and its relation to physics quite interesting** (information/enptropy, Boltmann distribution), **which made me reach a few interesting insigths**.

---
We will start with zero *class_confidence_level*.

Reminder: target discretizing - if the target relative price movement (%) is:
  * higher than confidence level: **up (go long) class (+1)**,
  * else if lower than the negative confidence level - **down (go short) class (-1)**,
  * else - **neutral class (0)**.

The classes directly represent the trading decision to be taken.

````
class_confidence_level=0
# ------------------ end of inputs ------------------

classification_obj=OzML.classification_shallow(
    split_data_obj=train_test_split_obj,class_confidence_level=class_confidence_level)

classification_obj.train_skLogisticRegression(
    multi_class='multinomial',solver='lbfgs',penalty='l2',C=1e5,n_jobs=1,max_iter=1000)
````

About the parameter choise:
* I chose 'multinomial' (=softmax) multi-class algorithm, rather than 'ovr' (one vs. the rest, performing $K$ independent binary logistic classifications for $K>2$ classes), to allow the class decisions to share information between them to reach better (efficient and richer) decisions. In softmax the class probabilities are dependent and there are only $K-1$ weigth vectors to learn for $K>2$ classes.
* **I chose *C=1e5* to have effectively no regularization, which is not interesting in this section.** Adding (non-negligible) regularization - adjusting the C parameter, should be done by cross validation.

Let's see the prediction:

````
classification_obj.predict()
classification_obj.analyze()
````

**It seems that the achieved accuracy (~1/2) is higher than a random guess accuracy (1/3 for 3 classes), but since *class_confidence_level=0*, effectively we have 2 classes** (the cases of having relative price movement of exactly 0, the neutral class (0), are negligible).

Notice that a RuntimeWarning was issued because there are no class (0) occurances in the training period, so a zero division was encountered in accuracy calculation for this class. **This is no error, and poses no problem in the calculations.**

We can examine the $percision=\frac{TP}{TP+FP}$ for each class:

````
classification_obj.precision_df
````

It seems that class (-1) over-fits more than class (1). We observe that indeed there are negligible class (0) occurances.

Let's have a closer look (the x axes are linked between subfigures, in an interactive figure  zoom is convenient but here we will need to re-plot):

# Deep learning (TensorFlow)
Until now I have presented and used shallow learning. My *regression_shallow* and *classification_shallow* classes inherit *regression* and *classification* classes, respectively. 

Similarly, *regression_deep* and *classification_deep* classes also inherit *regression* and *classification* classes, respectively, so **using deep rather than shallow learning with my classes is equivalent in terms of making predictions and trading decisions**.

## Regression NN
As I explained above, I constructed my classes such that *using* deep learning for regression, to make predictions and trading decisions, is not different than using linear regression. The difference is in the *learning*.

---
In order to train a neural network with my classes, the next two lines will suffice:


```
regression_obj=OzML.regerssion_deep(train_test_split_obj)
regression_obj.train_NN()
```

However, there are many hyper-parameters involved in training a neural network, so we will go through them:
* Gradient update will be done in batches:
  * ** *batch_size* ** is the number of data samples per gradient update.
  * ** *learning_rate* ** is the learning rate, should be tuned to have a well converging loss.
* I defined the training such that a training **epoch** is completing batch updates on all training samples. The maximal number of epochs until stopping training is ** *max_epochs* **.
* I defined a **callback** that stops the training when the validation loss has stopped improving on a validation period:
  * ** *validation_split* ** is the fraction of the training data to be used as validation data at the end of each epoch, selected from the end of the data.
  * ** *min_dloss_to_stop* ** is the minimum absolute change in the validation loss to qualify as an improvement. Since the loss is MSE (+regularization), the units are the target units squrated, $^2.
  * ** *training_patience* ** is the number of epochs with no improvement after which training will be stopped
* The **optimizer** can be defined externally, else GradientDescentOptimizer will be used in *regression_obj.train_NN*.
* The **neural network model architecture** can be defined externally, else a default architecture defined in *regression_obj.train_NN* will be used.

---
I chose parameters in advance to start with over-fitting, then tackle it:

````
# learning parameters
batch_size=round(0.2*len(train_test_split_obj.y_train_series))
learning_rate=1e-5
max_epochs=500

# callback parameters
validation_split=0.3
min_dloss_to_stop=0.5**2
training_patience=10

# optional- building optimizer externally
import tensorflow as tf
optimizer=tf.train.GradientDescentOptimizer(learning_rate=learning_rate)

# optional - building the NN model architecture externally
from tensorflow import keras

model=keras.Sequential()
model.add(keras.layers.Dense(50,
    input_shape=(len(train_test_split_obj.X_train_df.columns),),
                             activation='tanh'))
model.add(keras.layers.Dense(50, activation='relu'))
model.add(keras.layers.Dense(50, activation='relu'))
model.add(keras.layers.Dense(1))
# ------------------ end of inputs ------------------

regression_obj=OzML.regerssion_deep(train_test_split_obj)
regression_obj.train_NN(model,max_epochs=max_epochs,
    batch_size=batch_size,learning_rate=learning_rate,
    validation_split=validation_split,training_patience=training_patience,
    min_dloss_to_stop=min_dloss_to_stop,verbose=0)
regression_obj.plot_training()
````

We observe over-fitting in the learning curves, after the ankle (epoch ~42) excessive training decreased the training MSE, but increased the validation MSE.

Let's see it in the prediction:

````
regression_obj.predict()
regression_obj.plot_prediction(thick_linewidth=4,thin_linewidth=1,
                               rolling_window=0);
````

We observe how the prediction fits the training period, misses in the validation period (the last 30% of the original training data in blue, since we chose *validation_split=0.3*) and the test period.

---
Let's only simplify the network architecture (less neurons, less layers) and add regularization and dropout for demonstration:

````
model=keras.Sequential()
model.add(keras.layers.Dense(10,
    input_shape=(len(train_test_split_obj.X_train_df.columns),),
    activation='tanh'))
model.add(keras.layers.Dropout(rate=0.3,seed=1))
model.add(keras.layers.Dense(10, activation='relu'))
model.add(keras.layers.Dense(1,
                 kernel_regularizer=keras.regularizers.l1_l2(l1=0.1,l2=0.6)))
# ------------------ end of inputs ------------------

# training
regression_obj=OzML.regerssion_deep(train_test_split_obj)
regression_obj.train_NN(model,max_epochs=max_epochs,
    batch_size=batch_size,learning_rate=learning_rate,
    validation_split=validation_split,training_patience=training_patience,
    min_dloss_to_stop=min_dloss_to_stop,verbose=0)
regression_obj.plot_training()

# plotting prediction
regression_obj.predict()
regression_obj.plot_prediction(thick_linewidth=4,thin_linewidth=1,markersize=2,
                               rolling_window=0);
````

This model simplification reduced over-fitting. Still, linear regression had a (much) better fit.

**In order to get better results, the hyper-parameters should be optimized:**
* Learning / optimizer / callback parameters.
* Network architecture: layers, neurons in each layer, activations.
* Dropout, regularization.

## Cauality / data leakage verification by white noise regression

**Time series learning must maintain causality:** the target on a time $t'$, $y(t=t')$, can only be learnt/predicted by past information, $X(t<t')$.

To verify that data preperation and learning calculations do not violate causality (data  leakage), **I invented a simple verification - white noise regression**: regression on completely random inputs taken from a distribution $N(0,\sigma^2)$ must yield a completely random prediction, and therefore error distributed as $N(0,\sigma^2)$ - or else cheating with information from the future (=breaking causality)!

---
We will use linear regression for this purpose, since my *regression* class uses the regressor to make predictions in the same way for all learning algorithms - the only difference between different learinng algorithms is the training (which uses scikit learn / TensorFlow standard algorithms - we can assume they maintain causailty). 

````
white_y_std=5
# ------------------ end of inputs ------------------

# creating white noise from imported data in this project (data_df) 
(white_data_df,white_attribute_names_list,white_noise_target_name)=OzML.whiten(
    data_df,
    white_y_mean=50,white_y_std=white_y_std,white_X_mean=100,white_X_std=10,
    frequency='1H',seed=0)

# creating features from the white data
(white_X_df,white_y_series)=OzML.basic_feature_building(white_data_df,
    white_noise_target_name,
    white_attribute_names_list,include_target_history_as_features)

# splitting data, normalizing (by the same inputs we used in this project's splitting)
white_train_test_split_obj=OzML.train_test_split(white_noise_target_name,'% differences')
white_train_test_split_obj.split_by_dates(white_X_df,white_y_series,
    train_start_datetime=y_series.index[0],
    train_duration_years=6.5,test_duration_years=6.5)
white_train_test_split_obj.normalize()

# linear regression
white_regression_obj=OzML.regerssion_shallow(white_train_test_split_obj)
white_regression_obj.train_skLinearRegression()
white_regression_obj.predict()
ax1,ax2=white_regression_obj.plot_prediction(rolling_window=10000)
ax1.set_ylabel('$')
ax2.set_ylabel('$');
````

We set *white_y_std=5*, target white noise std of $\sigma=5$, and indeed got a completly random prediciton and error, with with error std of $\$4.986=0.997\sigma$ for the train, and $\$4.99=0.998\sigma$ for the test, as expected (the white data length is $N=111385\sim10^5$ so the relative error of the std should be $\approx1/\sqrt{N}\approx0.3\%$). **My codes maintain causality**.
